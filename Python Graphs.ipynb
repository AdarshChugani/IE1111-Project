```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
customers=pd.read_excel('kaya.xlsx',sheet_name='priorities')
customers.columns=['priority1','priority2','priority3','priority4','demographic']
gpage=customers.groupby('demographic')


ep1,ec1,eq1,sp1,sc1,sq1,se1,ac1,ae1,ap1,aq1=gpage['priority1'].value_counts().tolist()
ep2,ec2,ee2,eq2,sc2,sp2,sq2,se2,ac2,ae2,aq2,ap2=gpage['priority2'].value_counts().tolist()
eq3,ee3,ec3,ep3,sq3,sp3,sc3,se3,aq3,ae3,ap3,ac3=gpage['priority3'].value_counts().tolist()
ee4,ec4,eq4,se4,sc4,sp4,sq4,ap4,ae4,aq4,ac4=gpage['priority4'].value_counts().tolist()

elderly = {
  'Price':(ep1*4+ep2*3+ep3*2+0)/(ep1+ep2+ep3),
  'Environment/Ambience': (0*4+ee2*3+ee3*2+ee4)/(ee2+ee3+ee4),
  'Food Quality': (eq1*4+eq2*3+eq3*2+eq4)/(eq1+eq2+eq3+eq4),
  'Convenience': (ec1*4+ec2*3+ec3*2+ec4)/(ec1+ec2+ec3+ec4),
}

student = {
  'Price': (sp1*4+sp2*3+sp3*2+sp4)/(sp1+sp2+sp3+sp4),
  'Environment/Ambience': (se1*4+se2*3+se3*2+se4)/(se1+se2+se3+se4),
  'Food Quality': (sq1*4+sq2*3+sq3*2+sq4)/(sq1+sq2+sq3+sq4),
  'Convenience': (sc1*4+sc2*3+sc3*2+sc4)/(sc1+sc2+sc3+sc4),
}

adult = {
 'Price': (ap1*4+ap2*3+ap3*2+ap4)/(ap1+ap2+ap3+ap4),
  'Environment/Ambience': (ae1*4+ae2*3+ae3*2+ae4)/(ae1+ae2+ae3+ae4),
  'Food Quality': (aq1*4+aq2*3+aq3*2+aq4)/(aq1+aq2+aq3+aq4),
  'Convenience': (ac1*4+ac2*3+ac3*2+ac4)/(ac1+ac2+ac3+ac4),
}

df = pd.DataFrame([elderly,student, adult,])
priorities = list(df)
num_priority = 4
from math import pi

angles = [i / float(num_priority) * 2 * pi for i in range(num_priority)]
angles += angles[:1]  # repeat the first value to close the circle
import matplotlib.pyplot as plt

GRAY = '#000000'

# Clear the plot to start with a blank canvas.
plt.clf()

# Create subplots for each data series
series_1 = plt.subplot(1, 1, 1, polar=True)
series_2 = plt.subplot(1, 1, 1, polar=True)
series_3 = plt.subplot(1, 1, 1, polar=True)

# Draw one x-axis per variable and add labels
plt.xticks(angles, priorities, color=GRAY, size=8)

# Draw the y-axis labels. To keep the graph uncluttered,
# include lines and labels at only a few values.
plt.yticks(
  [1, 2, 3, 4],
  ['Lowest\n Priority\n', '', '', 'Highest\n Priority\n'],
  color='#001133',
  size=7
)

# Constrain y axis to range 0-100
plt.ylim(0,4)
series_1_values = df.loc[0] \
                    .values \
                    .flatten() \
                    .tolist()
series_1_values += series_1_values[:1]  # duplicate first element to close the circle

series_2_values = df.loc[1] \
                    .values \
                    .flatten() \
                    .tolist()
series_2_values += series_2_values[:1]   # duplicate first element to close the circle

series_3_values = df.loc[2] \
                    .values \
                    .flatten() \
                    .tolist()
series_3_values += series_3_values[:1] 

# Set up colors
ORANGE = '#FD7120'
BLUE = '#00BFFF'
GREEN='#1ffd23'

# Plot the first series
series_1.set_rlabel_position(0)
series_1.plot(
  angles,
  series_1_values,
  color=ORANGE,
  linestyle='solid',
  linewidth=1,
  label='Elderly',
)

# Plot the second series
series_2.set_rlabel_position(0)
series_2.plot(
  angles,
  series_2_values,
  color=BLUE,
  linestyle='solid',
  linewidth=1,
  label='Student',
)


series_3.set_rlabel_position(0)
series_3.plot(
  angles,
  series_3_values,
  color=GREEN,
  linestyle='solid',
  linewidth=1,
  label='Adult',
)
plt.legend(loc='upper right')
plt.savefig('aaaa.png',dpi=900,bbox_inches='tight')
```
